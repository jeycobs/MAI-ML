7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
full_sq_data = data[data['full_sq'] < 50]
res_data = pd.DataFrame(full_sq_data)
res_data['timestamp'] = pd.to_datetime(full_sq_data['timestamp'])
res_data = res_data[(res_data['timestamp'] >= '2011-01-01') & (res_data['timestamp'] <= '2014-12-31')]
# Добавляем столбец с кварталами
res_data['quarter'] = res_data['timestamp'].dt.to_period('Q')
median_price_quarter = res_data.groupby('quarter')['price_doc'].median().reset_index() #медианы цен по кварталам
median_price_quarter['quarter'] = median_price_quarter['quarter'].astype(str)
plt.figure(figsize=(10, 6))
sns.lineplot(data = median_price_quarter, x = 'quarter',y = 'price_doc',marker = 'o')
plt.grid()
plt.title('График изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года')
plt.xlabel('Кварталы')
plt.ylabel('Цены')
plt.xticks(rotation = 45)
plt.show()
````
